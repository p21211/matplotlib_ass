```python
# solution 2 :
# scatter plot in which line is not mode but only the point are marked according to coordinates.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# solution 3 : 
#     subplot function is used in order to draw mutiple plots more than in a figure.
#     it consists of the number of row and column and the third argument as the index of the plot.
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(1,4,1)
plt.plot(x1,y1)
plt.show()

plt.subplot(1,4,2)
plt.plot(x2,y2)
plt.show()

plt.subplot(1,4,3)
plt.plot(x3,y3)
plt.show()

plt.subplot(1,4,4)
plt.plot(x4,y4)
plt.show()


```

```python
# solution 4: 
# bar plot is used to prensent data in visual form to get it better, it present the categorical vaules. 
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.subplot(1,2,1)
plt.bar(company, profit)
plt.show()

plt.subplot(1,2,2)
plt.barh(company,profit)
plt.show()

```

```python
# solution 5 :
# A Box Plot is the visual representation of the statistical five number summary of a given data set
#     it diplays the characterstic of a distribution.
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
fig = plt.figure(figsize =(10, 7))
plt.subplot(1,2,1)
plt.boxplot(box1)
plt.show()
fig = plt.figure(figsize =(10, 7))
plt.subplot(1,2,2)
plt.boxplot(box2)
plt.show()



    
```
